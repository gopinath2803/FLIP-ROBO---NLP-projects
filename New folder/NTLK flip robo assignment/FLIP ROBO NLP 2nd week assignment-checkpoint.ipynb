```python
file = open(r'C:/Users/Fallen Angel/read.txt', encoding="utf-8").read()
file
```

```python
lower_case = file.lower()
lower_case
```

```python
# Creating my document into a library
dictionary = {file}
dictionary
```

```python
def cleaner(file):
    newString=re.sub(r"@[a-zA-Z0-9]+","",file)        
    newString=re.sub("#","",newString) 
    newString=' '.join([dictionary[t] if t in dictionary else t for t in newString.split(" ")])
    newString=re.sub(r'http\S+','',newString)         
    newString=re.sub(r"'s\b"," is",newString)         
    newString=re.sub(r"’s\b"," is",newString)
    newString=re.sub(r'[\d]','',newString)
    words=newString.split()                            
    return (" ".join(words)).strip()
```

```python
cleaned_file = cleaner(lower_case)
cleaned_file
```

```python
sent_tokenized_text =nltk.sent_tokenize(cleaned_file)
sent_tokenized_text
```

```python
sent_tokenized_word = [nltk.word_tokenize(sentence) for sentence in sent_tokenized_text]
sent_tokenized_word
```

```python
stop_words = ["i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself",
              "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself",
              "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these",
              "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do",
              "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while",
              "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before",
              "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again",
              "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each",
              "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than",
              "too", "very", "s", "t", "can", "will", "just", "don", "should", "now","new","nep","rs","shall","."]
```

```python
for i in range(len(sent_tokenized_word)):
    sent_tokenized_word[i] = [word for word in sent_tokenized_word[i]
                    if word not in stop_words]
sent_tokenized_word
```

```python

flatten_matrix = [] 
for sublist in sent_tokenized_word: 
    for val in sublist: 
        flatten_matrix.append(val)
print(len(flatten_matrix))
print(len(sent_tokenized_word))
```

```python
model = Word2Vec(sent_tokenized_word, min_count=1,size=100,sg=1)
words = model.wv.vocab
```

```python
# similar words
similar = model.wv.most_similar('college',topn=3)
similar
```

```python
# Visualizing Word2Vec 
college_file = ['biological','graduate','encountered','apple','idealistic','company','impression','finest','creation','macintosh','fired','beginner','freed','years','another','bought','creative','less','successful']
list_demo=[]
for i in college_file:
  words_list_similar = model.wv.most_similar(i,topn=5)
  #combine_words_list.append(words_list_similar)
  list_demo.extend(words_list_similar)
list_demo
```

```python
words = list(map(lambda x: x[0], list_demo))
len(words)
```

```python
def plot_w2v(w_list):
    z = model[w_list]
    pca = PCA(n_components=3)
    result = pca.fit_transform(z) 
    # create a scatter plot of the projection
    pyplot.scatter(result[:, 1], result[:, 2])
    for i, word in enumerate(w_list):
        pyplot.annotate(word, xy=(result[i, 0], result[i, 2]))
    pyplot.figure(figsize=(12,15))
    pyplot.show()
plot_w2v(words)
```
