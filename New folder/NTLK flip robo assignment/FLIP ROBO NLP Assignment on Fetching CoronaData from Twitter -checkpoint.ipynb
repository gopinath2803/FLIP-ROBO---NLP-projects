# Fetch one month of data of coronavirus from twitter before it's outbreak and after outbreak and compare peoples emotion

Importing required Libraries

```python
import nltk
import string
from collections import Counter
import matplotlib.pyplot as plt
import GetOldTweets3 as got
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from sklearn.externals import joblib
import pickle
```

```python
def get_tweets(keyword,start,end,mxm):
    tweet_criteria=got.manager.TweetCriteria().setQuerySearch(keyword).setSince(start).setUntil(end)\
    .setMaxTweets(mxm)
    tweets=got.manager.TweetManager.getTweets(tweet_criteria)
    text=[[tweet.text] for tweet in tweets]
    return text
```

```python
#Considering WHO's outbreak date 01 April 2020. Getting 30 days before outbreak
before=get_tweets('corona virus', '2020-03-01','2020-03-31',1000)
before
```

# Use NLTK library for tokenization and for stopwords

```python
text_before=''
for i in range(0,len(before)):
    text_before=text_before+'. '+before[i][0]
lower=text_before.lower()

cleaned_text=lower.translate(str.maketrans('','',string.punctuation))

from nltk.tokenize import word_tokenize
tokens=word_tokenize(cleaned_text,'english')
len(tokens)
```

```python
stop_words=stopwords.words('english')
from nltk.corpus import stopwords
words=[word for word in tokens if word not in stop_words]
len(words)
```

```python
lemma=[(WordNetLemmatizer().lemmatize(word)) for word in words]
len(lemma)
```

```python
clear_line=''
emotion_list=[]
with open('emotions.txt','r') as file:
    for line in file:
        clear_line=line.replace("\n",'').replace(" ","").replace(","," ").replace("'","").strip()
        word,emotion=clear_line.split(':')
        if word in lemma:
            emotion_list.append(emotion)
            
w=Counter(emotion_list)
print(w)
```

```python
fig,ax=plt.subplots(figsize=(10,6))
ax.bar(w.keys(),w.values())
fig.autofmt_xdate()
plt.show()
```

```python
sia=SentimentIntensityAnalyzer()
sia.polarity_scores(cleaned_text)
```

```python
cvect=CountVectorizer(stop_words=stopwords.words('english'))
cvect.fit_transform(A)
print(cvect.get_feature_names())
```

```python
cvect.vocabulary_
```

```python
cnew=cvect.transform(B)
cnew=cnew.toarray()
```

```python
model=pd.DataFrame(cnew,columns=cvect.get_feature_names())
model
```

```python
cnew.sum()
```

```python
cols=model.columns
dic={}
for col in cols:
    if model[col].sum()>=1:
        dic[col]=model[col].sum()
print(dic)
```
