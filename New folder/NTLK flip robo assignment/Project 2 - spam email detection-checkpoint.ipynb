# Importing libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('messages.csv')
df.head()
```

```python
email = df.copy()
email.drop(columns=['subject'],axis=1,inplace=True)
```

```python
email.head()
```

```python
email['label'].replace({0:'non-spam',1:'spam'},inplace=True)
```

```python
print ('non-spam and spam counts','\n',email.label.value_counts())
```

```python
#Ratio
print ('spam ratio = ', round(len(email[email['label']=='spam']) / len(email.label),2)*100,'%')
print ('non-spam ratio  = ', round(len(email[email['label']=='non-spam']) / len(email.label),2)*100,'%')
```

```python
#New column for Length of message
email['length'] = email.message.str.len()
email.head(5)
```

```python
# Convert all messages to lower case
email['message'] = email['message'].str.lower()
email.head()
```

```python
#Label coding 0 and 1
email['label'].replace({'non-spam':0,'spam':1},inplace=True)
email.head(40)
```

```python
email['message'][21]
```

```python
# Replace email addresses with 'email'
email['message'] = email['message'].str.replace(r'^.+@[^\.].*\.[a-z]{2,}$',
                                 'emailaddress')
# Replace URLs with 'webaddress'
email['message'] = email['message'].str.replace(r'^http\://[a-zA-Z0-9\-\.]+\.[a-zA-Z]{2,3}(/\S*)?$',
                                  'webaddress')
# Replace money symbols with 'moneysymb' (£ can by typed with ALT key + 156)
email['message'] = email['message'].str.replace(r'£|\$', 'dollers')
    
# Replace 10 digit phone numbers (formats include paranthesis, spaces, no spaces, dashes) with 'phonenumber'
email['message'] = email['message'].str.replace(r'^\(?[\d]{3}\)?[\s-]?[\d]{3}[\s-]?[\d]{4}$',
                                  'phonenumber')   
# Replace numbers with 'numbr'
email['message'] = email['message'].str.replace(r'\d+(\.\d+)?', 'numbr')
```

```python
email['message'][0]
```

```python
# Remove punctuation
email['message'] = email['message'].str.replace(r'[^\w\d\s]', ' ')

# Replace whitespace between terms with a single space
email['message'] = email['message'].str.replace(r'\s+', ' ')

# Remove leading and trailing whitespace
email['message'] = email['message'].str.replace(r'^\s+|\s+?$', '')
```

```python
email['message'][0]
```

```python
# Remove stopwords
import string
import nltk
from nltk.corpus import  stopwords
stop_words = set(stopwords.words('english') + ['u', 'ü', 'ur', '4', '2', 'im', 'dont', 'doin', 'ure'])
email['message'] = email['message'].apply(lambda x: ' '.join(
    term for term in x.split() if term not in stop_words))
```

```python
email['message'][0]
```

```python
# New column after puncuations,stopwords removal
emailcleaned = email.copy()
emailcleaned['clean_length'] = emailcleaned.message.str.len()
emailcleaned.head()
```

```python
# Total length removal
print ('Before Cleaning Length', emailcleaned.length.sum())
print ('After Cleaning Length', emailcleaned.clean_length.sum())
```

```python
# Message distribution BEFORE cleaning
f,ax = plt.subplots(1,2,figsize = (15,8))

sns.distplot(emailcleaned[emailcleaned['label']==1]['length'],bins=20,ax=ax[0],label='spam emails distribution',color='r')

ax[0].set_xlabel('Spam email length')
ax[0].legend()

sns.distplot(emailcleaned[emailcleaned['label']==0]['length'],bins=20,ax=ax[1],label='non-spam emails distribution')
ax[1].set_xlabel('non-spam email length')
ax[1].legend()

plt.show()
```

```python
# Message distribution AFTER cleaning
f,ax = plt.subplots(1,2,figsize = (15,8))

sns.distplot(emailcleaned[emailcleaned['label']==1]['clean_length'],bins=20,ax=ax[0],label='spam emails distribution',color='r')

ax[0].set_xlabel('Spam email length')
ax[0].legend()

sns.distplot(emailcleaned[emailcleaned['label']==0]['clean_length'],bins=20,ax=ax[1],label='non-spam emails distribution')
ax[1].set_xlabel('non-spam email length')
ax[1].legend()

plt.show()
```

```python
#Getting sense of loud words in spam 
from wordcloud import WordCloud

spams = emailcleaned['message'][emailcleaned['label']==1]

spam_cloud = WordCloud(width=700,height=500,background_color='white',max_words=50).generate(' '.join(spams))

plt.figure(figsize=(10,8),facecolor='r')
plt.imshow(spam_cloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

```python
#Getting sense of loud words in ham (non-spam)

ham = emailcleaned['message'][emailcleaned['label']==0]
spam_cloud = WordCloud(width=600,height=400,background_color='white',max_words=50).generate(' '.join(ham))
plt.figure(figsize=(10,8),facecolor='k')
plt.imshow(spam_cloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf_vec = TfidfVectorizer()
features_TFIDF = tf_vec.fit_transform(emailcleaned['message'])
```

```python
X = features_TFIDF
Y = emailcleaned['label']
```

```python
# Applying MultinomialNB Classifier
from sklearn.naive_bayes import MultinomialNB
NB_model = MultinomialNB()
```

```python
# Applying RandomForest Classifier
from sklearn.ensemble import RandomForestClassifier
RF_model=RandomForestClassifier(n_estimators=100,criterion='entropy')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state=42,shuffle = True,stratify = Y)
```

# Train the data

```python
NB_model.fit(X_train,Y_train)
```

```python
RF_model.fit(X_train, Y_train)
```

# Model Prediction

```python
#y_pred= NB_model.predict(X_test)
y_pred= RF_model.predict(X_test)
```

```python
y_pred
```

# Evaluating the models

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
#print ('Final score = > ', accuracy_score(Y_test,y_pred))   ## testing Accuracy

print("NB Training Accuracy:",NB_model.score(X_train,Y_train))
print("NB Testing Accuracy:",NB_model.score(X_test,Y_test))
print("    ")
print("RF Training Accuracy:",RF_model.score(X_train,Y_train))
print("RF Testing Accuracy:",RF_model.score(X_test,Y_test))
print("    ")
```

```python
Ya = Y_test  # Real Value
Yp = NB_model.predict(X_test) 

from sklearn.metrics import confusion_matrix
cm_NB = confusion_matrix(Ya,Yp)
print(cm_NB)

# For more clearence of classification show Heatmap
import seaborn as sb
f,ax = plt.subplots(figsize=(6,6))
sb.heatmap(cm_NB,annot=True,linewidths=5,linecolor='blue',fmt=".00f",ax=ax)     # fmt=".00f" -> Two decimal points with float
plt.show()

from sklearn.metrics import classification_report
class_report = classification_report(Ya,Yp)
print(class_report)
```

```python

Ya = Y_test  # Real Value
Yp = RF_model.predict(X_test) 

from sklearn.metrics import confusion_matrix
cm_RF = confusion_matrix(Ya,Yp)
print(cm_RF)

# For more clearence of classification show Heatmap
import seaborn as sb
f,ax = plt.subplots(figsize=(6,6))
sb.heatmap(cm_RF,annot=True,linewidths=5,linecolor='blue',fmt=".00f",ax=ax)     # fmt=".00f" -> Two decimal points with float
plt.show()

from sklearn.metrics import classification_report
class_report = classification_report(Ya,Yp)
print(class_report)
```
